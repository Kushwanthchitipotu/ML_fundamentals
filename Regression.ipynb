```python
#Understanding Error Surface
#All imports
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections
```

```python


########################################
#Generate meaningfull data
########################################
def height_weight_relationship(x):
    return 3.86 * x - 110.42

num_points=25

np.random.seed(42)  # for reproducibility
heights = np.random.uniform(50, 80, num_points)  # Random heights between 50 and 80 inches
noise = np.random.normal(0, np.sqrt(20), num_points)  # Gaussian noise with mean 0 and variance 20
weights=height_weight_relationship(heights)+noise


########################################
#Plot scatter plot of data
########################################
plt.scatter(heights,weights,label='Noisy Data')
plt.xlabel('Height (in inches)')
plt.ylabel('Weight (in pounds)')
plt.title('Scatter Plot of Height vs Weight with Noise')
plt.legend()
plt.grid(True)


########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################


#Sample a bunch of w's around w_opt and compute the associated error
w_opt=[-110.42,3.86]


w0_set=np.linspace(w_opt[0]-50,w_opt[0]+50,1000)
w1_set=np.linspace(w_opt[1]-50,w_opt[1]+50,1000)

error=np.zeros((len(w0_set), len(w1_set)))






#Compute the error
def Error(w,t,x): #inputs : 1)weight 2)data i.e (t,x)
    #Estimate the target
    X = np.column_stack((np.ones_like(x), x))
    est=np.dot(X,w)
    #Compute and return the error
    error=np.sum((t-est)**2)
    return error
for i, w0 in enumerate(w0_set):
    for j, w1 in enumerate(w1_set):
        w = np.array([w0, w1])
        error[i, j] = Error(w, weights, heights)



def least_squares_estimate(X, t):
    XTX_inv = np.linalg.inv(np.dot(X.T, X))
    w_opt = np.dot(np.dot(XTX_inv, X.T), t)
    return w_opt



#Plot 3D error surface and the corresponding contour plots
#Error surface is a function of w0 and w1

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(121, projection='3d')
w0_mesh, w1_mesh = np.meshgrid(w0_set, w1_set)
ax.plot_surface(w0_mesh, w1_mesh, error.T, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Error')
ax.set_title('3D Error Surface')

# Plot contour plots
plt.subplot(122)
contour_plot = plt.contour(w0_mesh, w1_mesh, error.T, levels=20, cmap='viridis')
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('Contour Plot of Error Surface')
plt.colorbar(contour_plot)

plt.show()

#Locate the minima of the error surface
min_index = np.unravel_index(np.argmin(error), error.shape)
w0_min, w1_min = w0_set[min_index[0]],w1_set[min_index[1]]



########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function

def LinearRegression(x,t): #inputs : 1)input data i.e (x). 2)target i.e (t)
    X = np.column_stack((np.ones_like(x), x))
    XTX_inv = np.linalg.inv(np.dot(X.T, X))
    w_opt = np.dot(np.dot(XTX_inv, X.T), t)
    return w_opt


#Estimate optimal weight's using "LinearRegression" function
W_opt=LinearRegression(heights,weights)



#Estimate the targets using the input x and the estimated weights
X = np.column_stack((np.ones_like(heights), heights))
est=np.dot(X,W_opt)
est1=np.dot(X,(w0_min,w1_min))
#Plot the estimated line on top of the above scatter plot

plt.scatter(heights,weights,label='Noisy Data')
plt.plot(heights,est,label='Estimated line least squared',color='red')
plt.plot(heights,est1,label='Estimated line surface optimization',color='green',linestyle='dashed')
plt.xlabel('Height (in inches)')
plt.ylabel('Weight (in pounds)')
plt.title('Scatter Plot of Height vs Weight with Noise')
plt.legend()
plt.grid(True)



########################################
#Compare the estimated weight's using least squares approach with the error surface approach
########################################
def comparison(a,b):
  return abs(a[0]-b[0])+abs(a[1]-b[1])


value1=comparison(w_opt,W_opt)
value2=comparison(w_opt,(w0_min,w1_min))
if value1>=value2:
  print('Surface optimization gives the better weights near to the original weights')
else:
  print('Least square gives better weight near to original weights')
```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################
def sin(x):
  return np.sin(2*np.pi*x)

def transform(x,order):
  k=0
  for i in range(0,order):
    k+=np.power(x,i)
  return k

def polynomial_features(x, degree):
    return np.power(x, np.arange(degree + 1))

# Function to calculate the least squares solution for a polynomial
def least_squares_polynomial(x, y, degree):
    X = polynomial_features(x, degree)
    w_opt = np.linalg.lstsq(X, y, rcond=None)[0]
    return w_opt

# Function to predict values using the polynomial model
def predict_polynomial(x, w_opt, degree):
    X = polynomial_features(x, degree)
    return np.dot(X, w_opt)

# Function to calculate root mean squared error
def calculate_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

num_points=20
x_set=np.linspace(0,1,num_points)
noise = np.random.normal(0, np.sqrt(0.01), num_points)
np.random.shuffle(x_set)
#Obtain train and test splits
train_x,test_x=x_set[:10],x_set[10:]
train_noise,test_noise=noise[:10],noise[10:]
train_y=sin(train_x)+train_noise
test_y=sin(test_x)+test_noise


########################################
#Fitting Mth degree polynomial using least squares approach
########################################
def PolynomialFit(X_train, Y_train, M, lamda):
    X = np.vander(X_train, M + 1, increasing=True)
    w_opt = np.linalg.solve(X.T @ X + lamda * np.identity(M + 1), X.T @ Y_train)
    return w_opt


#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    TrainError = np.vander(X_train, w_est.shape[0], increasing=True) @ w_est
    TestError = np.vander(X_test, w_est.shape[0], increasing=True) @ w_est
    #Return training and testing predictions
    return TrainError,TestError

#Complete the function
def PolynomialPred_Error(w_est,X_train,X_test,Y_train,Y_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    Y_train_pred,Y_test_pred=PolynomialPred(w_est,X_train,X_test)
    TrainError=np.sum((Y_train-Y_train_pred)**2)
    TestError=np.sum((Y_test-Y_test_pred)**2)
    #Return training and testing error
    return TrainError,TestError

#Iterate through range of M values
M_range=list(range(10))
TrError = []
TeError = []
train_pred = np.array([])
test_pred = np.array([])
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt=PolynomialFit(train_x,train_y,M,0)

    #Predict training and testing targets
    train_y_pred,test_y_pred=PolynomialPred(w_opt,train_x,test_x)
    train_pred = np.concatenate([train_pred, train_y_pred])
    test_pred = np.concatenate([test_pred, test_y_pred])


    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    trainerror,testerror=PolynomialPred_Error(w_opt,train_x,test_x,train_y,test_y)

    #Store them for plotting
    TrError.append(trainerror)
    TeError.append(testerror)



#Plot training and testing estimates alogwith the original targets
p = len(train_pred) // len(TrError)  # Use integer division to ensure p is an integer

plt.figure(figsize=(20,30 ))
for i in range(1, 11):
    start_index = int((i - 1) * p)
    end_index = int(i * p)
    plt.subplot(5,2, i)
    plt.scatter(train_x, train_y, label='Training Data')
    plt.scatter(test_x, test_y,color='orange', label='Test Data')

    # Add print statements to debug dimension
    plt.scatter(train_x, train_pred[start_index:end_index], color='red', label='Train Prediction')
    plt.scatter(train_x, test_pred[start_index:end_index], color='green', label='Test Prediction')
    plt.title(f'M = {i-1}')
    plt.legend()

plt.show()



#Plot training error vs polynomial order, and testing error vs polynomial order
fig = plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.plot(M_range,TrError,label='Training Error',marker='o')
plt.xlabel('polynomial order')
plt.ylabel('Training error')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(M_range,TeError,label='Testing Error',marker='o')
plt.xlabel('polynomial order')
plt.ylabel('Testing error')
plt.legend()
plt.grid(True)

plt.show()

########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################

# Increase the size of training data set to 100 points
num_points_large = 100
x_set_large = np.linspace(0, 1, num_points_large)
noise_large = np.random.normal(0, np.sqrt(0.01), num_points_large)
np.random.shuffle(x_set_large)
# Obtain train and test splits for the larger dataset
train_x_large, test_x_large = x_set_large[:50], x_set_large[50:]
train_noise_large, test_noise_large = noise_large[:50], noise_large[50:]
train_y_large = sin(train_x_large) + train_noise_large
test_y_large = sin(test_x_large) + test_noise_large

# Repeat the experiments with the larger dataset
M_range_large = list(range(10))
TrError_large = []
TeError_large = []

train_pred_large = np.array([])
test_pred_large = np.array([])

for M in M_range_large:
    # Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt_large = PolynomialFit(train_x_large, train_y_large, M, 0)

    # Predict training and testing targets
    train_y_pred_large, test_y_pred_large = PolynomialPred(w_opt_large, train_x_large, test_x_large)
    train_pred_large = np.concatenate([train_pred_large, train_y_pred_large])
    test_pred_large = np.concatenate([test_pred_large, test_y_pred_large])

    # Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    trainerror_large, testerror_large = PolynomialPred_Error(w_opt_large, train_x_large, test_x_large, train_y_large, test_y_large)

    # Store them for plotting
    TrError_large.append(trainerror_large)
    TeError_large.append(testerror_large)

# Plot training and testing estimates along with the original targets for the larger dataset
p_large = len(train_pred_large) // len(TrError_large)

plt.figure(figsize=(20, 30))
for i in range(1, 11):
    start_index_large = int((i - 1) * p_large)
    end_index_large = int(i * p_large)
    plt.subplot(5, 2, i)
    plt.scatter(train_x_large, train_y_large, label='Training Data')
    plt.scatter(test_x_large, test_y_large, color='orange', label='Test Data')

    # Add print statements to debug dimensions
    plt.scatter(train_x_large, train_pred_large[start_index_large:end_index_large], color='red', label='Train Prediction')
    plt.scatter(train_x_large, test_pred_large[start_index_large:end_index_large], color='green', label='Test Prediction')
    plt.title(f'M = {i - 1}')
    plt.legend()

plt.show()

# Plot training error vs polynomial order, and testing error vs polynomial order for the larger dataset
fig = plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.plot(M_range_large, TrError_large, label='Training Error', marker='o')
plt.xlabel('polynomial order')
plt.ylabel('Training error')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(M_range_large, TeError_large, label='Testing Error', marker='o')
plt.xlabel('polynomial order')
plt.ylabel('Testing error')
plt.legend()
plt.grid(True)

plt.show()




########################################
#Effect of regularization
########################################
#Consider a set of lambda's. For example: lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
#Repeat the experiments, i.e., plot the prediction and error in predictions with respect to model order. Contrast these results with those obtained without regularization.


# Effect of regularization
lamdas = [0, 1e-7, 1e-4, 1e-2, 1]
TrError_reg = []
TeError_reg = []

for lamda in lamdas:
    # Repeat the experiments with regularization
    TrError_reg_lamda = []
    TeError_reg_lamda = []

    for M in M_range_large:
        # Fit Mth order polynomial with regularization. Use the function "PolynomialFit"
        w_opt_reg = PolynomialFit(train_x_large, train_y_large, M, lamda)

        # Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        trainerror_reg, testerror_reg = PolynomialPred_Error(w_opt_reg, train_x_large, test_x_large, train_y_large, test_y_large)

        # Store them for plotting
        TrError_reg_lamda.append(trainerror_reg)
        TeError_reg_lamda.append(testerror_reg)

    TrError_reg.append(TrError_reg_lamda)
    TeError_reg.append(TeError_reg_lamda)

# Plot training and testing error with respect to model order for different lambda values
fig, axs = plt.subplots(len(lamdas), 1, figsize=(12, 8), sharex=True, sharey=True)

for i, lamda in enumerate(lamdas):
    axs[i].plot(M_range_large, TrError_reg[i], label=f'Training Error (lambda={lamda})', marker='o')
    axs[i].plot(M_range_large, TeError_reg[i], label=f'Testing Error (lambda={lamda})', marker='o')
    axs[i].set_title(f'Error vs Polynomial Order (lambda={lamda})')
    axs[i].legend()

plt.xlabel('polynomial order')
plt.ylabel('Error')
#plt.ylim(0, 1.5 * max(max(TrError_reg[-1]), max(TeError_reg[-1])))  # Adjust the y-axis limits for better visibility
plt.tight_layout()
plt.show()






########################################
#Effect of bias regularization
########################################
#Modify the function i.e include bias
#Generate data
train_y_large1=train_y_large+5
test_y_large1=test_y_large+5



#Estimate the polynomial with and without regularization constraint
TrError_large1 = []
TeError_large1= []

train_pred_large1 = np.array([])
test_pred_large1 = np.array([])

for M in M_range_large:
    # Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt_large1 = PolynomialFit(train_x_large, train_y_large1, M, 0)

    # Predict training and testing targets
    train_y_pred_large1, test_y_pred_large1 = PolynomialPred(w_opt_large1, train_x_large, test_x_large)
    train_pred_large1 = np.concatenate([train_pred_large1, train_y_pred_large1])
    test_pred_large1 = np.concatenate([test_pred_large1, test_y_pred_large1])

    # Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    trainerror_large1, testerror_large1 = PolynomialPred_Error(w_opt_large1, train_x_large, test_x_large, train_y_large1, test_y_large1)

    # Store them for plotting
    TrError_large1.append(trainerror_large1)
    TeError_large1.append(testerror_large1)



#Compare the two estimated polynomials and report the observations
x_axis=np.linspace(0,1,1000)
X_points=np.vander(x_axis, w_opt_large1.shape[0], increasing=True)
old=X_points@w_opt_large
new=X_points@w_opt_large1

plt.figure(figsize=(10, 6))
plt.plot(x_axis, old, label='Old estimated polynomial', linestyle='--', color='blue')
plt.plot(x_axis, new, label='New estimated polynomial', linestyle='-', color='red')


plt.xlabel('x-axis')
plt.ylabel('values')
plt.title('Comparison of with and without bias')
plt.legend()


plt.show()
```
